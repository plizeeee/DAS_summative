```python
### load the meta data
from datetime import datetime
start = datetime.now()
data = []
with gzip.open('meta_Grocery_and_Gourmet_Food.json.gz') as f:
    for l in f:
        data.append(json.loads(l.strip()))
print(datetime.now()-start)
# total length of list, this number equals total number of products
print(len(data))

# first row of the list
print(data[0])
```

```python
# data[1]
from collections import Counter
# Counter(data[0]['category'])

i = 0
for row in data:
  if i == 0:
    cts = Counter(row['category'])
  else:
    cts = cts + Counter(row['category'])
  i +=1
```

```python
i = 0
lst_asins = list()
for row in data:
  if 'Cooking & Baking' in row['category']:
    lst_asins.append(row['asin'])
```

```python
import pandas as pd
import gzip
import json
from datetime import datetime

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

start = datetime.now()
df_foods = getDF('Grocery_and_Gourmet_Food_5.json.gz')
print(datetime.now()-start)
```
